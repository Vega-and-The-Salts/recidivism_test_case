```python
dataTrain = pd.read_csv('train.csv')
```

```python
sns.set(rc={'figure.figsize':(25,15)})
sns.scatterplot(x=dataTrain.Id, y=dataTrain.SalePrice, size=dataTrain.SalePrice, hue=dataTrain.OverallCond, style=dataTrain.YrSold, sizes=(60,300), palette="magma")
plt.title('Scatterplot of raw data')
plt.show()
```

```python
df_corrs = dataTrain.corr()
good_corrs = df_corrs.index[abs(df_corrs["SalePrice"]) > 0.5]
X_train, X_test, y_train, y_test = train_test_split(dataTrain[good_corrs].iloc[:, :-1], dataTrain[good_corrs].iloc[:, -1], test_size=.4, random_state=10)
```

```python
model = RandomForestClassifier(max_depth=6)
model.fit(X_train,y_train)
y_pred = model.predict(X_train)
print(f"Random Forest Classification Model score: {round(r2_score(y_train,y_pred)*100,2)}")
```
